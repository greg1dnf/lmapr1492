```python
from IPython.display import Image
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("uV0oAcDvINnFw7CK") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-961711")
```

```python
xrdcalculator = XRDCalculator(1.54060, 1)
pattern = xrdcalculator.get_pattern(structure)
```

```python
print("Premier pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[0])
print("Intensité :", pattern.y[0])
print("Indices hkl :", pattern.hkls[0])
print("Valeur de d :", pattern.d_hkls[0])
```

```python
print("Deuxième pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[1])
print("Intensité :", pattern.y[1])
print("Indices hkl :", pattern.hkls[1])
print("Valeur de d :", pattern.d_hkls[1])
```

```python
print("Troisième pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[2])
print("Intensité :", pattern.y[2])
print("Indices hkl :", pattern.hkls[2])
print("Valeur de d :", pattern.d_hkls[2])
```

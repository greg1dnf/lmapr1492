```python
from IPython.display import Image
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
with MPRester("uV0oAcDvINnFw7CK") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-961711")
```

```python
symétries = SpacegroupAnalyzer(structure).get_symmetry_operations()
#print(symétries)

#Symétries choisies
S1 = symétries[1]
S2 = symétries[2]
S3 = symétries[8]


#Atomes 

Zr = [0.5, 0.5, 0.5]
Si = [0, 0, 0]
Pt = [0.25, 0.25, 0.25]


#Application des symétries 

Zr_Sym = S1.operate(Zr)
Si_Sym = S2.operate(Si)
Pt_Sym = S3.operate(Pt)

```

```python
print("Opération 4-barre, inversion centre en 0 0 0 (y,z,-x-y-z),sur l'atome Zr : \n")
print(" L'opérateur de symetrie 4-barre est celui-ci:", "\n" , S1,"\n")

print("L'atome de départ est le Zr et a comme coordonnées :", Zr, "\n")
print("L'Atome après la symétrie a comme coordonées : ",Zr_Sym,"\n")


view = quick_view(structure)
display(view)
view.script("draw symop 2 {atomno=1}")

Image(filename='Zr_Sym.png')

```

```python
print("Opération de symétrie C2, (z,-x-y-z,x) ,sur l'atome Si : \n")
print(" L'opérateur de d'inversion C2 est celui-ci:", "\n" , S2,"\n")

print("L'atome de départ est le Si et a comme coordonnées :", Si, "\n")
print("L'Atome après la symétrie a comme coordonées : ",Si_Sym,"\n")

view = quick_view(structure)
display(view)
view.script("draw symop 3 {atomno=2}")

Image(filename='Si_Sym.png')
```

```python
print("Opération de symétrie C3, (-x-y-z,x,z) ,sur l'atome Pt : \n")
print(" L'opérateur de plan miroir est celui-ci:", "\n" , S3,"\n")

print("L'atome de départ est le Si et a comme coordonnées :", Pt, "\n")
print("L'Atome après la symétrie a comme coordonées : ",Pt_Sym,"\n")

view = quick_view(structure)
display(view)
view.script("draw symop 9 {atomno=3}")

Image(filename='Pt_Sym.png')
```
